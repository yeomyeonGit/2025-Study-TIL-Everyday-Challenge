```python
# str 접근자
# 문자열 메서드를 활용하기 위한 도구
df["이름"].str
```

```python
# 알파벳 대/소문자 변환
# .str.upper() / .str.lower()
# 이메일 데이터를 소문자로 변환
# 데이터프레임을 수정 X
df["이메일"] = df["이메일"].str.lower()
df
```

```python
# 특정 문자열이 포함되어있는지 확인
# str.contains("문자열")
# 이메일에서 ".error" 포함 여부를 확인
condition = df["이메일"].str.contains(".error")

# 데이터 필터링
df.loc[condition]
```

```python
# 문자열을 특정 문자열로 대체
# str.replace("기존 문자열", "대체 문자열")
# 이메일의 ".error" 를 ".com" 대체
# 데이터프레임을 수정 X

df["이메일"] = df["이메일"].str.replace(".error", ".com")
df
```

```python
# 특정 문자열로 시작 / 끝 확인
# str.startswith("문자열") / str.endswith("문자열")
# 성별에서 "남" 으로 시작하는 데이터 확인
condition = df["성별"].str.startswith("남")

# 데이터 필터링
df.loc[condition]
```

```python
# 특정 문자열을 기준으로 문자열을 분할
# str.split("특정 문자열")
# 이메일을 "@" 를 기준으로 분할
df["이메일"].str.split("@")

# 이메일을 "@" 를 기준으로 분할한 후
# 첫 번째 값을 추출
df["아이디"] = df["이메일"].str.split("@").str[0]
df

# 두 번째 값을 추출
df["도메인"] = df["이메일"].str.split("@").str[1]
df
```

```python
# 공백 제거
# str.strip() 양쪽 공백을 제거
# str.lstrip() 왼쪽(시작) 공백을 제거
# str.rstrip() 오른쪽(끝) 공백을 제거
# 이름에서 양쪽 공백을 제거
# 데이터프레임을 수정 X
df["이름"] = df["이름"].str.strip()
df
```

```python
# 문자열의 길이 계산
# str.len()
# 이름의 길이(문자) 계산
# 데이터프레임을 수정 X
df["이름"].str.len()
```
