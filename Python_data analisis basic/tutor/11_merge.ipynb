```python
# 11_merge.ipynb
import pandas as pd

df_left = pd.DataFrame(
    {
        "Key": ["K0", "K1", "K2", "K3"],
        "A": ["A0", "A1", "A2", "A3"],
        "B": ["B0", "B1", "B2", "B3"],
    }
)

df_right = pd.DataFrame(
    {
        "Key": ["K0", "K1", "K2", "K4"],
        "C": ["C0", "C1", "C2", "C3"],
        "D": ["D0", "D1", "D2", "D3"],
    }
)
```

```python
df_left.head()
```

```python
df_right.head()
```

```python
# inner join
pd.merge(
    df_left,
    df_right,
    on="Key",  # 병합 기준 열
    how="inner",  # 병합 방식
)
```

```python
pd.merge(
    df_left,
    df_right,
    on="Key",
    how="left",
)
```

```python
import pandas as pd
import random

# 컴퓨터에서 실제로 random 이란건 존재할 수 없다.
# 시간(초)

# seed : ?
# random 모듈의 결과 고정시키는 값
random.seed(0)

# 학생 데이터프레임
student_df = pd.DataFrame(
    {
        "학생ID": range(101, 111),
        "학과": [
            random.choice(["컴퓨터공학과", "전자공학과", "기계공학과"])
            for _ in range(10)
        ],
    }
)

# 강의 데이터프레임
lecture_df = pd.DataFrame(
    {
        "강의ID": [1001, 1002, 1003, 1004],
        "강의 이름": ["프로그래밍 기초", "웹 개발", "데이터 분석", "데이터베이스"],
        # "교수 ID":[],
        # "강의 장소"
    }
)

n = 100
# 수강 데이터프레임
enrollment_df = pd.DataFrame(
    {
        "수강ID": range(1, n + 1),
        "학생ID": [random.choice(student_df["학생ID"]) for _ in range(n)],
        "강의ID": [random.choice(lecture_df["강의ID"]) for _ in range(n)],
        "학점": [random.choice(["A", "B", "C", "D", "F"]) for _ in range(n)],
    }
)
```

```python
student_df.head()
```

```python
lecture_df.head()
```

```python
enrollment_df.head()
```

```python
pd.merge(student_df, enrollment_df, on="학생ID", how="inner").head(20)
```

```python
pd.merge(enrollment_df, lecture_df, on="강의ID", how="left").head(20)
```
