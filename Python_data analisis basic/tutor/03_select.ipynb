```python
# 레이블 기반 선택 .loc[]
# .loc[행, 열]

# 모든 행을 선택
df.loc[:]

# 행 이름이 "0" 인 행을 선택
df.loc[0]

# # 행 이름이 "0", "1", "3" 인 행을 선택
# df.loc[0, 1, 3]

# # 슬라이싱, 종료 레이블이 포함
df.loc[0:10]

# # 단일 열(name) 선택
df.loc[:, "name"]

# # 복수 열(name,seller_type) 선택
df.loc[:, ["name", "seller_type"]]

# # 행과 열 선택 조합
df.loc[[0, 3, 5], ["name", "selling_price"]]
df.loc[0:10, ["name", "selling_price"]]
```

```python
# 1. selling_price 열 기준으로 내림차순으로 정렬해서
# 2. 결과를 새로운 변수(sorted_df)에 저장하세요.
sorted_df = df.sort_values(by="selling_price", ascending=False)
sorted_df
# 3. sorted_df 에 대해서 인덱스 기반 선택을 하세요.
# .iloc[0]

# 4. sorted_df 에 대해서 레이블 기반 선택을 하세요.
# .loc[0]
```

```python
sorted_df.iloc[0]  # 위치 기반
```

```python
# 조건식
# 열을 선택해서 해당 열의 개별 데이터와 값을 비교하는 식
# 판매 가격(selling_price) 100000 보다 작은

# 1. 열 선택
# 2. 조건 값을 작성
# 3. 비교연산자를 작성

condition = df["selling_price"] < 100000

# 4. loc를 사용해서 데이터를 선택(필터링)
df.loc[condition]
```

```python
# 주행거리(km_driven)가 100000 보다 큰 데이터를 필터링
# 1. 열 선택
# 2. 조건 값을 작성
# 3. 비교연산자를 작성
# 4. loc를 사용해서 데이터를 선택(필터링)
condition = df["km_driven"] > 100000
# 기타. name열과 selling_price 열을 선택
df.loc[condition, ["name", "selling_price"]]
```

```python
# 연료(fuel)가 "Diesel" 인 데이터를 필터링
# 1. 열 선택
# 2. 조건 값을 작성
# 3. 비교연산자를 작성
# 4. loc를 사용해서 데이터를 선택(필터링)

condition = df["fuel"] == "Diesel"
df.loc[condition]
```

```python
# 판매 유형(seller_type)이 "Dealer" 가 아닌 데이터를 필터링
# 1. 열 선택
# 2. 조건 값을 작성
# 3. 비교연산자를 작성
# 4. loc를 사용해서 데이터를 선택(필터링)
condition = df["seller_type"] != "Dealer"
df.loc[condition]
```

```python
# 주행거리가 100000 보다 작은
condition1 = df["km_driven"] < 100000

# 판매가격이 100000 보다 작은
condition2 = df["selling_price"] < 100000

# 두 조건을 모두 만족하는 데이터를 선택
df.loc[condition1 & condition2]
```

```python
# 주행거리가 100000 보다 작은
condition1 = df["km_driven"] < 100000

# 판매가격이 100000 보다 작은
condition2 = df["selling_price"] < 100000

# 운전방식(transmission)이 Automatic 인 차량
condition3 = df["transmission"] == "Automatic"

# 위 세 조건을 모두 만족하는 데이터
df.loc[condition1 & condition2 & condition3]
```

```python
# fuel 가 Electric 이거나 transmission이 Automatic 인 차량을 선택

condition1 = df["fuel"] == "Electric"
condition2 = df["transmission"] == "Automatic"

df.loc[condition1 | condition2]
```

```python
# year가 2018 보다 커야하고
# fuel가 Petrol 또는 Diesel 인 데이터를 선택
condition1 = df["year"] > 2018
condition2 = df["fuel"] == "Petrol"
condition3 = df["fuel"] == "Diesel"

df.loc[condition1 & (condition2 | condition3)]
```

```python
# 연료(fuel)가 Petrol, Diesel, LPG 인 데이터를 탐색
condition = df["fuel"].isin(["Petrol", "Diesel", "LPG"])

# 연료(fuel)가 Petrol, Diesel, LPG 인 데이터를 선택
df.loc[condition]

# 연료(fuel)가 Petrol, Diesel, LPG 이 아닌 데이터를 선택
# ~ 를 통해 False 인 데이터를 선택
df.loc[~condition]
```

```python
# year가 2020, 2018, 2016, 2014 인 데이터를 선택
condition = df["year"].isin([2020, 2018, 2016, 2014])
df.loc[condition]
```
