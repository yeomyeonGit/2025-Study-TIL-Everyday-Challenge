* 딕셔너리로 데이터프레임을 만드는 이유

딕셔너리로 데이터를 정의하는 것은 Pandas의 데이터프레임을 만들 때 가장 직관적이고 일반적인 방법 중 하나입니다. 그 이유는 다음과 같습니다:

* 딕셔너리 구조는 데이터프레임과 자연스럽게 매핑된다.
딕셔너리는 키(key)와 값(value)으로 구성된 데이터 구조입니다.


* 확장성과 유연성이 높다.
새 열을 추가하거나 데이터를 수정하기 쉬움:

data["grade"] = ["A", "B", "C"]  # 'grade' 열 추가

여러 데이터 타입(문자, 숫자 등)을 혼합하여 처리할 수 있어 다양한 데이터를 다루기에 적합합니다.

* 다른 데이터 소스와 호환성: 
JSON, API 응답, 파이썬 코드에서 다루는 데이터는 보통 딕셔너리 형태로 제공됩니다.
Pandas는 딕셔너리를 자연스럽게 데이터프레임으로 변환할 수 있어 데이터를 다루기 편리합니다.

```python
# 리스트 활용 데이터프레임 생성 코드
# 리스트 안에 딕셔너리

data = [
    {"name": "민수", "age": 16},
    {"name": "지연", "age": 19},
    {"name": "현우", "age": 19},
    {"name": "수진", "age": 16, "location": "서울"},
]

df = pd.DataFrame(data)

print(df)

# none 없다
# NaN 없다. 결측치
```
