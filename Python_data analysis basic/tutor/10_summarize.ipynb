```python
# 평균 연봉이 6000 이상인 부서만 필터링
# 1. 그룹화
grouped = df.groupby("부서")


# 2. 조건 함수
# 매개변수 : 부서를 기준으로 그룹화된 행 데이터
def greater_salary(group):
    # 평균 연봉
    mean_salary = group["연봉"].mean()

    # 평균 연봉이 6000이상이면 return True
    if mean_salary >= 6000:
        return True

    # 아니라면 return False
    else:
        return False


# 3. filter() 적용
grouped.filter(greater_salary)
```

```python
# 평균 연봉이 6000 이상인 부서만 필터링
# 1. 그룹화
# group_rows : 그룹화된 데이터 행의 모음
group_rows = df.groupby("부서")


# 2. 조건함수
# group_row : 그룹화된 데이터의 개별 행
def greater_salary(group_row):
    mean_salary = group_row["연봉"].mean()

    return mean_salary >= 6000


# 3. 필터링
group_rows.filter(greater_salary)
```

```python
# 연봉 열에 평균, 중위값
# 적용하고 싶은 집계 함수를 리스트에 문자열로 입력
# mean, median
df["연봉"].agg(["mean", "median"])

# 나이 열의 최소, 최대, 평균
df["나이"].agg(["min", "max", "mean"])
```

```python
# 여러개의 열에 여러개의 집계함수
"""
딕셔너리(key-value)를 활용해서 열과 열에 적용할 집계 함수 작성
key : 열
value : 집계함수

Dataframe.agg(
    {   
        "열 1": "min",
        "열 2": ["min","max"]
    }
)
"""
# 이름의 개수
# 연봉의 평균, 중위값
# 나이의 최소, 최대, 평균
df.agg(
    {
        "이름": "count",
        "연봉": ["mean", "median"],
        "나이": ["min", "max", "mean"],
    }
)
```

```python
# 그룹화된 데이터에도 다중 집계 함수 agg()
grouped = df.groupby("지역")

result = grouped.agg(
    {
        "이름": "count",
        "연봉": ["mean", "median"],
        "나이": ["min", "max", "mean"],
    }
)
# 다중 집계함수의 결과 : 컬럼들이 계층 구조를 가진다.
# 만약에 연봉 - mean 선택하고 싶으면?
result[("연봉", "mean")]

# 나이 - min, max 선택
result.loc[:, ("나이", ["min", "max"])]
```

```python
df = pd.read_csv("vgsales.csv")
df.head()
```

```python
# 장르별 세계 판매량 합계를 출력한다.
group_rows = df.groupby("genre")
group_rows["total_sales"].sum().sort_values(ascending=False)
```

```python
# 플랫폼별 세계 판매량 평균
group_rows = df.groupby("console")

# 가장 평균이 높은 플랫폼의 값
group_rows["total_sales"].mean().sort_values(ascending=False).iloc[0]
```

```python
# 개발사별 게임 출시 개수
group_rows = df.groupby("developer")

# 게임 출시 개수가 가장 많은 개발사(행 이름)
group_rows["title"].count().sort_values(ascending=False).index[1]
```

```python
# 플랫폼 그룹화
group_rows = df.groupby("console")

# 일본 판매량 합계를 기준으로 정렬
group_rows.agg(
    {
        "total_sales": ["sum", "mean"],
        "na_sales": ["sum", "mean"],
        "jp_sales": ["sum", "mean"],
    }
).sort_values(by=("jp_sales", "sum"), ascending=False)
# jp_sales 열과 sum 열은 계층구조이기 때문에
# 소괄호 () 를 사용해서 계층 열을 표현한다.
```

```python
drop_df = df.dropna(axis="index", subset=["critic_score"])
# 평론가 평점이 9.5가 넘는 개발사 데이터
group_rows = drop_df.groupby("developer")


def greater_score(group):
    # 그룹의 평균 평론가 평점이 9.5 보다 크면
    # True 를 반환, 아니면 False 를 반환
    # 즉, 평균 평점이 9.5보다 큰 데이터만 선택
    return group["critic_score"].mean() > 9.5


group_rows.filter(greater_score)


# [1,1,1,1,NaN]
# (1+1+1+1) / 4
```
