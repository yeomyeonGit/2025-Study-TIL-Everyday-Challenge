```python
# 04_operate.ipynb
import pandas as pd

df = pd.DataFrame(
    {
        "A": [1, 2, 3],
        "B": [4, 5, 6],
    },
)
```

```python
# 행 이름이 0의 A 열 데이터 수정
df.loc[0, "A"] = 0
df
```

```python
# 행 이름이 0, 1의 B 열 데이터 수정
df.loc[[0, 1], "B"] = 1
df
```

```python
# 행 이름이 0, 1의 B 열 데이터 수정
df.loc[[0, 1], "B"] = [10, 20]
df
```

```python
# A 열의 데이터를 모두 10으로 수정
df.loc[:, "A"] = 10
df
```

```python
# B열의 데이터 중 10보다 작은 데이터를 10으로 수정
# 1. 10보다 작은 B열의 데이터를 필터링
condition = df["B"] < 10

# 2. 필터링한 후 값 수정
df.loc[condition, "B"] = 10
df
```

```python
# 1. 이상치를 수정
# 이상치란, 키가 평균 173cm 누군가 키가 300cm 누군가 키가 10cm

# 2. 값의 통일성을 위해 수정
# 지역 열이 있는데 어떤 데이터는 서울, 어떤 데이터는 서울특별시, 어떤 데이터는 서울시
```

```python
# 열 'A' 이름을 '가' 로 수정
# 기존 데이터프레임을 수정하지 않는다.
df = df.rename(columns={"A": "가"})
df
```

```python
# 행 이름 '0'을 '1'로 수정
df = df.rename(index={0: 1})

# 행 이름 '1'의 수가 2개
df

# 행 이름 '1'을 선택?
df.loc[1]
```

```python
# 데이터 타입을 수정
df = pd.DataFrame({"A": [1, 2, 3], "B": [4.5, 5.5, 6.4]})

# astype()
# 기존 데이터프레임을 수정하지 않는다.
df["A"] = df["A"].astype(float)
print(df.info())
df
```

```python
# 행 추가하기
# concat() : 2개 이상의 데이터프레임을 합치는 함수

df = pd.DataFrame({"A": [1, 2, 3], "B": [4, 5, 6]})
new_df = pd.DataFrame({"A": [7], "B": [8]})

# ignore_index : 합친 후 행 이름을 초기화한다.
df = pd.concat([df, new_df], ignore_index=True)
df
```

```python
# 만약에 존재하지 않는 열에 대해 값을 저장하면?
# 새로운 열을 생성하고, 값을 저장
df["C"] = 10
df
```

```python
# B열의 데이터가 짝수(2로 나눴을 때 나머지가 0)라면
condition = df["B"] % 2 == 0

df["D"] = condition
df
```

```python
import numpy as np

condition = df["B"] % 2 == 0
df["E"] = np.where(condition, "짝수", "홀수")
```

```python
# 중복 데이터 삭제
unique_df = df.drop_duplicates()
```

```python
unique_df = unique_df.drop(
    columns=[
        "alive",
        "embarked",
        "class",
        "deck",
        "who",
    ]
)
```

```python
# 요금이 0인 행을 선택
condition = unique_df["fare"] == 0

# 요금이 0인 행의 "fare" 열의 값을 10으로 수정
unique_df.loc[condition, "fare"] = 10
```

```python
# 0인 데이터가 존재하는지 확인
unique_df.sort_values(by="fare", ascending=True)
```

```python
# 나이가 1보다 작은 행을 선택
condition = unique_df["age"] < 1

# 나이가 1보다 작은 행의 "age" 열 데이터를 1로 수정
unique_df.loc[condition, "age"] = 1
```

```python
# 나이가 1보다 작은 행이 존재하는지 확인
unique_df.sort_values(by=["age"])
```
